```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
rent_apartment_url = "https://raw.githubusercontent.com/5x12/course-prodml/refs/heads/main/module1/1_model_creation_jupyter_notebook/rent_apartments.csv"
data = pd.read_csv(rent_apartment_url)
```

```python
data.head()
```

2. Data preparation

```python
data.dtypes
```

```python
data_encoded = pd.get_dummies(data, columns=["balcony", "parking", "furnished", "garage", "storage"], drop_first=True)
```

```python
data_encoded
```

```python
data_encoded.garden.unique()
```

```python
import re
```

```python
data_encoded.garden
```

```python
int(re.findall(r'\d+', data_encoded.garden[4])[0])
```

```python
len(data_encoded)
```

```python
for i in range(len(data_encoded.garden)):
    if data_encoded.garden[i] == "Not present":
        data_encoded.garden[i] = 0
    else:
        data_encoded.garden[i] = int(re.findall(r'\d+', data_encoded.garden[i])[0])
```

```python
data_encoded.garden.unique()
```

# 3. Model Building

```python
data_encoded.head()
```

```python
data_encoded.columns.tolist()
```

```python
X = data_encoded[["area", "constraction_year", "bedrooms", "garden", "balcony_yes", "parking_yes", "furnished_yes", "garage_yes", "storage_yes"]]
```

```python
X.head()
```

```python
y = data_encoded.rent
```

```python
y
```

# Split the dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Model Building

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```

```python
rf.predict([[85, 2015, 2, 20, 1,1, 0, 0, 1]])
```
